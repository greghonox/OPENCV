```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('px-people.jpg')
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
```

```python
classfication = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces = classfication.detectMultiScale (img_gray, 1.3, 5)
len(faces)
```

```python
plt.figure(figsize=(20, 10)) # configure size of the plot
for x, y, w, h in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 0), 3)
    
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
face_img = 0
for x, y, w, h in faces:
    face_slice = img[y:y+h, x:x+w]
    img_roi = cv2.vtColor(face_slice, cv2.COLOR_BGR2RGB)
    cv2.imsave(f'/tmp/face_{face_img}.jpg', img_roi)
    print(f'save img: face_{face_img}.jpg')
```
